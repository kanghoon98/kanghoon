## What Is Machine Learning?

기계 학습은 컴퓨터 프로그래밍의 과학 (및 예술)으로 데이터에서 배울 수 있습니다.
다음은 조금 더 일반적인 정의입니다.

기계 학습은 컴퓨터가 명시 적으로 프로그래밍 되지 않고 학습 할 수 있는 학습 영역입니다.
-Arthur Samuel, 1959

그리고 엔지니어링 지향적 인 것 :

컴퓨터 프로그램은 P에 의해 측정된 T에서의 성능이 경험 E로 향상되면 일부 작업 T와 성능 측정 P에 대한 경험 E로부터 학습이라고 합니다.
-Tom Mitchell, 1997

예를 들어, 스팸 필터는 기계 학습 프로그램으로, 
스팸 전자 메일 및 일반 전자 메일의 예를 들어 스팸 플래그를 배울 수 있습니다.
시스템이 학습에 사용하는 예제를 학습 세트라고 합니다. 
각 교육 예제를 교육 샘플이라고 합니다. 
이 경우 작업 T는 새 이메일에 대해 스팸 플래그를 지정하고, 경험 E는 학습 데이터이며 성능 측정 값 P를 정의해야 합니다. 
예를 들어, 올바르게 분류된 전자 메일의 비율을 사용할 수 있습니다.
이 특정 성능 측정 값은 정확도라고 하며 분류 작업에 자주 사용됩니다
위키 백과 사본을 다운로드하면 컴퓨터에 더 많은 데이터가 있지만 갑자기 모든 작업에서 더 나은 것은 아닙니다. 
따라서 기계 학습이 아닙니다.

## Why Use Machine Learning?

기존 프로그래밍 기법을 사용하여 스팸 필터를 작성하는 방법을 고려하십시오

1)먼저 스팸이 일반적으로 어떻게 보이는지 살펴봅니다. "4U", "신용 카드", "무료"및 "놀라운"과 같은 일부 단어나 문구가 주제에 많이 등장하는 경향이 있음을 알 수 있습니다. 아마도 보낸 사람의 이름, 전자 메일의 본문 등에 몇 가지 다른 패턴이 있음을 알 수 있습니다.

2)발견 한 각각의 패턴에 대해 탐지 알고리즘을 작성하고, 이러한 패턴이 여러 개 발견되면 프로그램에서 전자 메일을 스팸으로 표시합니다.

3)프로그램을 테스트하고 1 단계와 2 단계를 반복하여 충분히 좋은 상태가 될 때까지 반복하십시오.

문제는 사소한 것이 아니기 때문에 프로그램은 복잡한 규칙의 긴 목록이 될 가능성이 있습니다.
대조적으로, 기계 학습 기술을 기반으로 한 스팸 필터는 스팸 예제에서 비정상적으로 빈번한 단어 패턴을 햄 예제 (그림 1-2)와 비교하여 감지함으로써 스팸의 좋은 예측 인자 인 단어 및 구문을 자동으로 학습합니다. 이 프로그램은 훨씬 더 짧고, 유지 보수가 쉬우며, 정확할 가능성이 큽니다.


게다가 스팸 발송자가 "4U"가 포함된 모든 이메일을 차단 한 사실을 알게 되면 "For U"라고 쓰게됩니다. 전통적인 프로그래밍 기법을 사용하는 스팸 필터는 "For U"이메일 플래그로 업데이트해야 합니다. 스팸 발송자가 스팸 필터를 계속 사용한다면 새로운 규칙을 계속 작성해야 합니다.
반대로 Machine Learning 기술을 기반으로 한 스팸 필터는 "For U"가 사용자가 신고 한 스팸이 비정상적으로 빈번하게 발견되어 사용자의 개입없이 스팸을 신고하기 시작합니다 (그림 1-3).


Machine Learning이 주로 사용되는 또 다른 영역은 기존 접근 방식에 비해 너무 복잡하거나 알려진 알고리즘이 없는 영역입니다. 예를 들어, 음성 인식을 고려해보십시오.
간단하게 시작하여 "one"과 "two"라는 단어를 구별할 수 있는 프로그램을 작성하려고 합니다. "2"라는 단어가 높은 음높이 ( "T" ), 높은 피치 사운드 강도를 측정하는 알고리즘을 하드 코딩하고 이를 사용하여 1과 2를 구별 할 수 있습니다.
이 기술은 시끄러운 환경과 수십 개의 언어로 수백만 명의 매우 다른 사람들이 말하는 수천 개의 단어로 확장되지 않습니다. 가장 좋은 해결책은 각 단어에 대해 많은 예제 녹음이 주어지면서 스스로 배우는 알고리즘을 작성하는 것입니다.

마침내 Machine Learning은 인간이 학습하는 데 도움을 줄 수 있습니다 (그림 1-4).
ML 알고리즘은 그들이 배운 것을 보기 위해 검사 할 수 있습니다 (일부 알고리즘의 경우 까다로울 수 있음). 예를 들어 스팸 필터가 충분한 스팸에 대해 교육을 받으면 쉽게 검사하여 스팸의 가장 좋은 예측 자라고 믿는 단어 목록과 단어 조합을 표시 할 수 있습니다. 때로는 예상치 못한 상관 관계 나 새로운 추세가 드러나고 이로 인해 문제를 더 잘 이해하게됩니다.
ML 기법을 적용하여 많은 양의 데이터를 파헤쳐 보면 즉시 나타나지 않은 패턴을 발견하는 데 도움이됩니다. 이를 데이터 마이닝이라고합니다.

요약하면, 기계 학습은 다음과 같은 경우에 유용합니다.

1. 기존 솔루션에 많은 수작업 튜닝 또는 긴 규칙 목록이 필요한 문제 : 한 가지 학습 알고리즘으로 코드를 단순화하고 성능을 향상시킬 수 있습니다.

3. 변화하는 환경 : 기계 학습 시스템은 새로운 데이터에 적응할 수 있습니다.

4. 복잡한 문제 및 많은 양의 데이터에 대한 통찰력 확보.

## Types of Machine Learning Systems

다양한 종류의 기계 학습 시스템이 있으므로 다음을 기반으로 광범위한 카테고리로 분류하는 것이 유용합니다.
- 그들이 사람에의해 감독 (감독, 감독되지 않음, 반 감독, 강화 학습)으로 훈련 받았는지 여부
- 즉석에서 점진적으로 학습 할 수 있는지 여부 (온라인 대 일괄 학습)
- 과학자들이하는 것처럼 새로운 데이터 포인트를 알려진 데이터 포인트와 간단히 비교하거나 훈련 데이터의 패턴을 감지하고 예측 모델을 구축하여 작업하는지 (인스턴스 기반 대 모델 기반 학습)

이 기준은 배타적이지 않습니다. 당신은 당신이 좋아하는 어떤 방법으로 그들을 결합 할 수 있습니다. 예를 들어 최첨단 스팸 필터는 스팸 및 햄의 예를 사용하여 교육 된 심층 신경 네트워크 모델을 사용하여 실시간으로 학습 할 수 있습니다. 온라인 모델 기반의 감독 학습 시스템입니다.

이 기준들 각각을 조금 더 자세히 살펴 보겠습니다.

Supervised/Unsupervised Learning

기계 학습 시스템은 훈련 중에받는 감독의 양과 유형에 따라 분류 할 수 있습니다. 감독 학습, 감독되지 않은 학습, 반 감독 학습, 강화 학습의 네 가지 주요 범주가 있습니다.

감독 학습에서 알고리즘에 제공하는 교육 데이터에는 레이블이라는 원하는 솔루션이 포함됩니다 (그림 1-5).

일반적인 감독 학습 과제는 분류입니다. 
스팸 필터는 이에 대한 좋은 예입니다. 스팸 필터는 클래스 (스팸 또는 햄)와 함께 많은 예제 이메일과 함께 학습되며 새 이메일을 분류하는 방법을 알아야합니다.
또 다른 일반적인 작업은 예측자라고 불리는 일련의 기능 (주행 거리, 나이, 브랜드 등)이 주어지면 자동차 가격과 같은 목표 수치를 예측하는 것입니다. 이런 종류의 일을 회귀 (regression)라고 부릅니다. (그림 1-6) 
시스템을 교육하기 위해서는 예측 변수와 라벨 (예 : 가격)을 포함하여 자동차에 대한 많은 예제를 제공해야합니다.

note

기계 학습에서 속성은 데이터 유형 (예 : '마일리지')이지만, 기능은 문맥에 따라 여러 가지 의미가 있습니다. 일반적으로 속성과 값 (예 : '주행 거리 = 15,000')을 의미합니다. 그러나 많은 사람들이 단어 속성과 기능을 서로 바꾸어 사용할 수 있습니다.

일부 회귀 알고리즘도 분류에 사용할 수 있으며 그 반대도 마찬가지입니다. 예를 들어, 로지스틱 회귀는 분류에 일반적으로 사용되며, 주어진 클래스에 속할 확률에 해당하는 값을 출력 할 수 있습니다 (예 : 스팸 일 확률 20 %).

다음은 이 책에서 다루는 가장 중요한 감독 학습 알고리즘입니다.
- k- 가장 가까운 이웃
- 선형 회귀
- 로지스틱 회귀
- 지원 벡터 머신 (SVM)
- 의사 결정 트리 및 임의의 포리스트
- 신경 회로망 

감독되지 않은 학습

감독되지 않은 학습에서는 추측 할 수 있듯이 교육 데이터의 레이블이 지정되지 않습니다 (그림 1-7). 체계는 교사없이 배우려고합니다.

다음은 가장 중요한 감독되지 않은 학습 알고리즘입니다 (8 장의 차원 감소를 설명합니다).

클러스터링
- 케이 - 수단
- 계층 적 클러스터 분석 (HCA)
- 기대 극대화


시각화 및 차원 감소
- 주성분 분석 (PCA)
- 커널 PCA
- 로컬 선형 임베딩 (LLE)
- t- 분산 확률 적 이웃 내장 (t-SNE)


연관 규칙 학습
- 선험적으로
- 명성

예를 들어 블로그 방문자에 대한 많은 데이터가 있다고 가정 해보십시오. 클러스터링 알고리즘을 실행하여 유사한 방문자 그룹을 탐지하려고 할 수 있습니다 (그림 1-8). 어느 시점에 방문자가 속한 그룹인지 알고리즘에 알리지는 않습니다. 당신의 도움없이 이러한 연결을 찾습니다. 예를 들어, 40 %의 방문자가 만화를 좋아하고 저녁 시간에 블로그를 읽는 반면 20 %는 주말에 방문하는 젊은 SF 사용자입니다. 계층 적 클러스터링 알고리즘을 사용하면 각 그룹을 더 작은 그룹으로 세분 할 수도 있습니다. 이렇게하면 각 그룹의 게시물을 타겟팅하는 데 도움이됩니다.

시각화 알고리즘은 감독되지 않은 학습 알고리즘의 좋은 예이기도합니다. 
복잡하고 레이블이없는 많은 데이터를 제공하고 쉽게 플롯 할 수있는 데이터의 2D 또는 3D 표현을 출력합니다 (그림 1-9). 이러한 알고리즘은 가능한 한 많은 구조를 유지하려고합니다 (예 : 입력 공간에서 별도의 클러스터를 시각화에서 겹치지 않도록 유지). 그러면 데이터가 어떻게 구성되어 있는지 이해하고 의심하지 않은 패턴을 식별 할 수 있습니다.

TIP

다른 기계 학습 알고리즘 (예 : 감독 학습 알고리즘)에 입력하기 전에 치수 감소 알고리즘을 사용하여 교육 데이터의 크기를 줄이는 것이 좋습니다. 훨씬 빨라지고 데이터가 디스크와 메모리 공간을 덜 차지하게되며 경우에 따라 더 나은 성능을 발휘할 수도 있습니다.

또 다른 중요한 감독되지 않는 작업은-예를 들어, 사기를 방지하기 위해 비정상적인 신용 카드 거래를 감지하거나, 제조 결함을 포착하거나, 다른 학습 알고리즘에 공급하기 전에 데이터 세트에서 특이 값을 자동으로 제거하는 등의 비정상 탐지입니다. 이 시스템은 정상적인 인스턴스로 교육을 받았고 새로운 인스턴스를 볼 때 정상적인 인스턴스인지 여부 또는 비정상 가능성 여부를 알 수 있습니다 (그림 1-10 참조).

마지막으로, 감독되지 않은 또 다른 작업은 연관 규칙 학습입니다.
목표는 많은 양의 데이터를 파헤 치고 특성 간의 흥미로운 관계를 발견하는 것입니다.
예를 들어 슈퍼마켓을 소유하고 있다고 가정합니다. 영업 로그에 연관 규칙을 적용하면 바베큐 소스와 감자 칩을 구매하는 사람들도 스테이크를 사기 쉽다는 것을 알 수 있습니다. 따라서이 항목들을 서로 가깝게 배치 할 수 있습니다. 

반 감독 학습

일부 알고리즘은 부분적으로 레이블이 지정된 교육 데이터, 일반적으로 많은 레이블이 지정되지 않은 데이터와 약간의 레이블이 지정된 데이터를 처리 할 수 있습니다. 이것을 반 감독 학습이라고합니다 (그림 1-11).
Google 포토와 같은 일부 사진 호스팅 서비스가 이에 해당됩니다. 모든 가족 사진을 서비스에 업로드하면 같은 사람 A가 사진 1, 5 및 11에 나타나고 다른 사람 B가 사진 2, 5 및 7에 나타남을 자동으로 인식합니다. 이것은 알고리즘의 감독되지 않은 부분입니다 (클러스터링). 이제 모든 시스템 요구 사항은 당신이 이 사람들이 누구인지를 말해주는 것입니다. 일인당 단 하나의 레이블, 모든 사진의 모든 사람들에게 이름을 지으면 사진 검색에 유용합니다.

대부분의 반 감독 학습 알고리즘은 감독되지 않은 알고리즘과 감독 된 알고리즘의 조합입니다. 예를 들어, deep belief networks (DBN)는 제한된 볼츠만 기계 (RBM)라고 불리는 감독되지 않은 구성 요소를 기반으로합니다. RBM은 감독되지 않은 방식으로 순차적으로 훈련되고 전체 시스템은 감독 된 학습 기술을 사용하여 미세 조정됩니다.

강화학습

강화 학습은 에이전트라고 불리는 학습 시스템은 환경을 관찰하고, 행동을 선택하고 수행하며, 대가로 보상을받습니다 (또는 그림 1-12에서 보듯이 부정적인 보상의 형태로 처벌받을 수 있음). 그런 다음 시간이 지남에 따라 가장 많은 보상을 얻기 위해 정책이라고하는 최상의 전략을 스스로 학습합니다. 정책은 상담원이 주어진 상황에있을 때 선택해야하는 작업을 정의합니다.

예를 들어, 많은 로봇이 도보 학습 방법을 배우기 위해 강화 학습 알고리즘을 구현합니다. DeepMind의 AlphaGo 프로그램은 강화 학습의 좋은 예이기도합니다. 2017 년 5 월에 세계 챔피언 Ke Jie를 Go 게임에서 이길 때 헤드 라인을 장식했습니다. 수백만 개의 게임을 분석 한 다음 자체적으로 여러 게임을 실행하여 성공한 정책을 배웠습니다. 챔피언과의 경기에서 학습이 중단되었음을 유의하십시오. AlphaGo는 방금 배운 정책을 적용하고있었습니다.

Batch and Online Learning

기계 학습 시스템을 분류하는 데 사용되는 또 다른 기준은 시스템이 들어오는 데이터 스트림에서 점진적으로 학습 할 수 있는지 여부입니다.


일괄 학습

배치 학습(일괄 학습)에서 시스템은 점진적으로 학습 할 수 없으며 모든 사용 가능한 데이터를 사용하여 교육을 받아야합니다. 일반적으로 많은 시간과 컴퓨팅 리소스가 필요하므로 일반적으로 오프라인으로 처리됩니다. 먼저 시스템이 훈련되고 생산 단계로 넘어 가서 더 이상 학습하지 않고 실행됩니다(그것은 단지 배운 것을 적용합니다). 이를 오프라인 학습이라고합니다.

일괄 학습 시스템이 새로운 유형의 스팸과 같은 새로운 데이터에 대해 알기를 원하면 새로운 데이터뿐만 아니라 이전 데이터까지 전체 데이터 세트의 처음부터 새 버전의 시스템을 교육해야합니다. 그런 다음 이전 시스템을 중지하고 새 시스템으로 교체하십시오.
다행히도, 기계 학습 시스템의 전체 교육, 평가 및 시작 프로세스는 매우 쉽게 자동화 될 수 있으므로 (그림 1-3 참조) 일괄 학습 시스템조차도 변경에 적응할 수 있습니다. 간단히 데이터를 업데이트하고 필요에 따라 시스템의 새 버전을 처음부터 교육하십시오.

이 솔루션은 간단하고 종종 정상적으로 작동하지만 전체 데이터 세트를 사용하는 교육에는 많은 시간이 걸릴 수 있으므로 일반적으로 매 24 시간마다 또는 매주 새로운 시스템을 교육하게됩니다. 시스템이 빠르게 변화하는 데이터 (예 : 주가 예측)에 적응해야하는 경우보다 효과적인 해결책이 필요합니다.
또한 전체 데이터 집합을 학습하려면 많은 컴퓨팅 리소스 (CPU, 메모리 공간, 디스크 공간, 디스크 I / O, 네트워크 I / O 등)가 필요합니다. 많은 양의 데이터가 있고 매일 시스템을 처음부터 교육하도록 자동화하면 결국 많은 비용이 들게됩니다. 데이터 양이 많으면 일괄 학습 알고리즘을 사용하는 것이 불가능할 수도 있습니다.
마지막으로, 시스템이 자율적으로 학습 할 수 있어야하고 제한된 리소스 (예 : 스마트 폰 응용 프로그램 또는 화성 탐사선)가있는 경우 다량의 교육 데이터를 들고 많은 시간 동안 교육을위한 많은 리소스를 사용합니다.
다행히도 이러한 모든 경우에 더 나은 옵션은 점진적으로 학습 할 수있는 알고리즘을 사용하는 것입니다.

온라인 학습

온라인 학습에서는 개별적으로 또는 미니 배치라는 작은 그룹별로 순차적으로 데이터 인스턴스를 제공하여 점진적으로 시스템을 교육합니다. 각 학습 단계는 빠르고 저렴하므로 새로운 데이터가 도착하자마자 시스템에서이를 알 수 있습니다 (그림 1-13 참조).

온라인 학습은 데이터를 연속 흐름 (예 : 주가)으로 수신하고 신속하게 또는 자율적으로 변경해야하는 시스템에 적합합니다. 또한 컴퓨팅 리소스가 제한되어있는 경우에도 좋은 옵션입니다. 온라인 학습 시스템에서 새 데이터 인스턴스에 대해 알게되면 더 이상 필요하지 않으므로 삭제할 수 있습니다 (이전으로 롤백 할 수 없으면 데이터를 "재생"합니다). 이렇게하면 엄청난 양의 공간을 절약 할 수 있습니다.

온라인 학습 알고리즘을 사용하여 하나의 시스템의 주요 메모리에 맞지 않는 거대한 데이터 세트의 시스템을 교육 할 수도 있습니다 (이는 코어 외 학습이라고 함). 알고리즘은 데이터의 일부를로드하고 해당 데이터에 대해 학습 단계를 실행하며 모든 데이터에서 실행될 때까지 프로세스를 반복합니다 (그림 1-14 참조).

주의

이 전체 과정은 일반적으로 오프라인으로 (즉, 실제 시스템이 아닌) 수행되므로 온라인 학습은 혼란스런 이름이 될 수 있습니다. 점진적 학습이라고 생각하십시오.

온라인 학습 시스템의 한 가지 중요한 매개 변수는 변화하는 데이터에 얼마나 빨리 적응해야 하는지를 뜻하는 학습 속도입니다. 높은 학습 속도를 설정하면 시스템이 새로운 데이터에 빠르게 적응하지만 이전 데이터를 빠르게 잊어 버리는 경향이 있습니다 (스팸 필터가 표시된 최신 유형의 스팸 만 표시하는 것을 원치 않는 경우).반대로 낮은 학습 속도를 설정하면 시스템의 관성이 커집니다. 즉, 더 느리게 익숙해 지지만 새로운 데이터 또는 비 대표적인 데이터 요소의 시퀀스에 대한 노이즈에 덜 민감합니다.

온라인 학습의 가장 큰 문제점은 잘못된 데이터가 시스템에 공급되면 시스템의 성능이 점차 저하된다는 것입니다.
이러한 위험을 줄이려면 시스템을 면밀히 모니터링해야하며 성능 저하를 감지하면 즉시 학습을 중단하고 이전 작업 상태로 되돌릴 수 있습니다. 또한 입력 데이터를 모니터링하고 비정상 데이터에 반응 할 수 있습니다 (예 : 변형 탐지 알고리즘 사용).

Instance-Based Versus Model-Based Learning

기계 학습 시스템을 분류하는 또 다른 방법은 일반화하는 방법입니다. 대부분의 기계 학습 과제는 예측에 관한 것입니다. 이것은 여러 가지 훈련 예제가 주어지면 이전에 본 적이없는 예를 일반화 할 수 있어야 함을 의미합니다. 훈련 자료에 대한 좋은 성과의 척도는 좋지만 충분하지는 않습니다. 진정한 목표는 새로운 인스턴스에서 잘 수행하는 것입니다.

일반화에는 두 가지 주요 접근법인 인스턴스 기반 학습과 모델 기반 학습이 있습니다.

인스턴스 기반 학습

아마도 학습의 가장 사소한 형태는 단순히 마음으로 배우는 것입니다. 이 방법으로 스팸 필터를 만드는 경우 사용자가 이미 신고한 이메일과 동일한 모든 이메일을 플래그로 표시합니다. 최악의 해결책은 아니지만 확실하지는 않습니다.

알려진 스팸 전자 메일과 동일한 전자 메일을 플래그하는 대신 스팸 필터는 알려진 스팸 전자 메일과 매우 유사한 전자 메일을 플래그하도록 프로그래밍 할 수 있습니다. 이를 위해서는 2 개의 전자 메일 간의 유사성 측정이 필요합니다. 두 이메일 간의 (매우 기본적인) 유사성 척도는 공통된 단어의 수를 세는 것일 수 있습니다. 알려진 전자 메일과 공통된 단어가 많은 경우 시스템에서 전자 메일을 스팸으로 플래그 지정합니다.

이것을 인스턴스 기반 학습이라고합니다. 시스템은 예제를 핵심으로 학습한 다음 유사성 척도 (그림 1-15)를 사용하여 새로운 사례로 일반화합니다.

모델 기반 학습

일련의 예제에서 일반화하는 또 다른 방법은 이러한 예제의 모델을 작성한 다음 해당 모델을 사용하여 예측을 만드는 것입니다. 이를 모델 기반 학습이라고합니다 (그림 1-16).

예를 들어, 돈으로 사람들이 행복하게되는지 알고 싶다고 가정하면 OECD 웹 사이트의 Better Life Index 데이터와 IMF 웹 사이트에서 1 인당 GDP 통계를 다운로드하십시오. 그런 다음 테이블에 가입하고 1 인당 GDP를 기준으로 정렬합니다. 표 1-1은 여러분이 얻은 것을 발췌 한 것입니다.

임의의 국가에 대한 데이터를 그래프로 표현해보자 (그림 1-17).

추세를 살펴보면 국가의 1 인당 GDP가 증가함에 따라 삶의 만족도가 어느 정도 선형적으로 상승한 것처럼 보입니다. 따라서 1 인당 GDP의 선형 함수로서 삶의 만족도를 모델링하기로 결정했습니다. 이 단계를 모델 선택이라고합니다. 한 속성 인 1 인당 GDP (식 1-1)만으로 삶의 만족에 대한 선형 모델을 선택했습니다.

이 모델에는 두 가지 모델 매개 변수 인 θ0과 θ1.5가 있습니다.이 매개 변수를 조정하면 그림 1-18과 같이 모델이 선형 함수를 나타낼 수 있습니다.

모델을 사용하려면 매개 변수 값 θ0 및 θ1을 정의해야합니다. 모델을 가장 잘 수행 할 수있는 값을 어떻게 알 수 있습니까? 이 질문에 대답하려면 성능 측정 값을 지정해야합니다. 모델이 얼마나 좋은지를 측정하는 유틸리티 함수 (또는 피트니스 함수)를 정의하거나 모델의 품질을 측정하는 비용 함수를 정의 할 수 있습니다. 선형 회귀 문제의 경우 일반적으로 사람들은 선형 모델의 예측과 학습 예제 간의 거리를 측정하는 비용 함수를 사용합니다. 목표는 이 거리를 최소화하는 것입니다.

이것은 선형 회귀 알고리즘이 나오는 곳입니다 : 당신은 당신의 훈련 예제를 제공하고 선형 모델이 당신의 데이터에 가장 잘 맞도록하는 매개 변수를 찾습니다. 이를 모델 교육이라고합니다. 이 경우 알고리즘은 최적의 매개 변수 값이 θ0 = 4.85이고 θ1 = 4.91 × 10^-5임을 알게됩니다.
이제 모델은 그림 1-19에서 볼 수 있듯이 가능한 한 가장 근접한 교육 데이터에 맞 춥니 다 (선형 모델의 경우).

마침내 예측을하기 위해 모델을 실행할 준비가되었습니다. 예를 들어 키프로스가 얼마나 행복하고 OECD 데이터에 해답이 없는지 알고 싶다고 가정 해보십시오. 다행히도 모델을 사용하여 1 인당 GDP를 조사하고 22,587 달러를 찾고 나서 모델을 적용하면 수명 만족도는 4.85 + 22,587 × 4.91 × 10-5 = 5.96가 됩니다.

예제 1-1에서는 데이터를로드하고 준비하며 시각화를위한 산점도를 만든 다음 선형 모델을 훈련하고 예측을하는 Python 코드를 보여줍니다.

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model

# Load the data
oecd_bli = pd.read_csv("oecd_bli_2015.csv", thousands=',')
gdp_per_capita = pd.read_csv("gdp_per_capita.csv",thousands=',',delimiter='\t',
                             encoding='latin1', na_values="n/a")

# Prepare the data
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]

# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]

```

모든 것이 잘 진행되면 모델이 좋은 예측을 할 것입니다. 그렇지 않은 경우 더 많은 특성 (고용률, 건강, 대기 오염 등)을 사용하거나, 더 우수한 품질의 교육 데이터를 얻거나,보다 강력한 모델 (예 : 다항 회귀 모델)을 선택해야 할 수 있습니다.

## Main Challenges of Machine Learning

간단히 말해서, 당신의 주된 임무는 학습 알고리즘을 선택하고 그것을 일부 데이터로 훈련하는 것이므로 잘못 될 수있는 두 가지는 "잘못된 알고리즘"과 "불량 데이터"입니다. 불량 데이터의 예부터 시작해 봅시다.

Insufficient Quantity of Training Data(교육 자료의 불충분한 양)

유아가 사과가 무엇인지 알기 위해서는 사과를 가리키고 "사과"라고 말하면됩니다 (아마도이 과정을 몇 번 반복하면됩니다). 이제 아이는 모든 종류의 색상과 모양의 사과를 인식 할 수 있습니다. (천재).
기계 학습은 아직 그다지 빠르지 않습니다. 대부분의 기계 학습 알고리즘이 제대로 작동하려면 많은 양의 데이터가 필요합니다. 아주 간단한 문제 일지라도 수천 개의 예제가 필요하며 이미지 또는 음성 인식과 같은 복잡한 문제의 경우 수백만 가지 예제가 필요할 수 있습니다 (기존 모델의 일부를 다시 사용할 수있는 경우 제외).

Nonrepresentative Training Data(비 대표성 교육 데이터)

일반화하기 위해서는 훈련 데이터가 일반화하려는 새로운 사례를 대표하는 것이 중요합니다. 이는 인스턴스 기반 학습이나 모델 기반 학습을 사용하든 상관 없습니다.

예를 들어 선형 모델을 훈련하기 위해 이전에 사용한 국가 세트는 완벽하게 대표되지 않았습니다.(몇몇 국가는 표기하지 않았었습니다). 그림 1-21은 누락 된 국가를 추가 할 때 데이터가 어떻게 보이는지 보여줍니다.

이 데이터에서 선형 모델을 학습하면 실선이되고 구 모델은 점선으로 표시됩니다. 보시다시피, 누락 된 일부 국가를 추가하면 모델이 크게 변경 될뿐만 아니라 그러한 단순한 선형 모델이 절대로 잘 작동하지 않을 것임을 분명히합니다. 매우 부유 한 국가들은 보통 부유 한 국가보다 행복하지 않으며 (사실 그들은 불행한 것처럼 보입니다) 반대로 일부 빈국은 많은 부유 한 국가보다 행복하게 보입니다.

대표성이없는 교육 세트를 사용하여, 특히 빈곤국 및 매우 부유 한 국가의 경우 정확한 예측을 내릴 수없는 모델을 교육했습니다.

일반화하려는 사례를 대표하는 교육 세트를 사용하는 것이 중요합니다. 샘플이 너무 작 으면 샘플링 노이즈 (예 : 우연한 결과로 표현할 수없는 데이터)를 갖지만 샘플링 방법에 결함이있는 경우 매우 큰 샘플조차도 비현실적일 수 있습니다. 이를 샘플링 바이어스라고합니다.

Poor-Quality Data(품질이 낮은 데이터)

물론 교육 데이터에 오류, 이상 치 및 노이즈 (예 : 품질이 좋지 않은 측정 결과)가 가득하면 시스템이 기본 패턴을 감지하기가 더 어려워 지므로 시스템 성능이 떨어집니다. 교육 데이터를 정리하는 데 많은 시간을 할애할 가치가 있습니다. 사실, 대부분의 데이터 과학자들은 그렇게하는 데 많은 시간을 소비합니다.

-일부 인스턴스가 분명히 특이 치일 경우,이를 단순히 버리거나 수동으로 오류를 수정하는 것이 도움이 될 수 있습니다.

-일부 인스턴스에서 몇 가지 기능이 누락 된 경우 (예 : 5 %의 고객이 연령을 지정하지 않은 경우)이 속성을 모두 무시할지, 이러한 인스턴스를 무시할지, 누락 된 값을 채울지 (예 : 중간 값 나이), 또는 하나의 모델과 그 모델이없는 모델 하나 등을 훈련시킬 수 있습니다.

Irrelevant Features(관련없는 기능)

교육 자료에 관련성이 충분하고 관련성이없는 자료가 너무 많지 않은 경우에만 시스템에서 학습 할 수 있습니다. Machine Learning 프로젝트의 성공에 결정적으로 중요한 부분은 훈련 할 수있는 좋은 기능들이 있습니다. 피쳐 엔지니어링 (feature engineering)이라고하는 이 프로세스에는 다음이 포함됩니다.

-피쳐 선택 : 기존 피쳐들 중에서 훈련 할 가장 유용한 피쳐를 선택합니다.

-피쳐 추출 : 기존 피쳐를 결합하여보다 유용한 피쳐를 생성합니다 (앞서 살펴본 바와 같이 치수 감소 알고리즘이 도움이됩니다).

-새로운 데이터를 수집하여 새로운 피쳐를 생성합니다.

이제 나쁜 데이터의 많은 예를 살펴 보았습니다. 잘못된 알고리즘의 몇 가지 예를 살펴 보겠습니다.

Overfitting the Training Data(교육 자료 과적합)

외국을 방문 중이며 택시 기사가 당신에게 사기를 친다고 가정 해보십시오. 당신은 그 나라의 모든 택시 기사가 도둑이라고 말하고 싶은 유혹을 느낄 것입니다. 지나친 일반화는 우리 인간이 너무 자주하는 일이며, 불행히도 조심하지 않으면 기계가 같은 함정에 빠질 수 있습니다. 기계 학습에서는 이것을 overfitting이라고합니다. 모델이 교육 데이터에서 잘 수행되지만 잘 일반화되지는 않습니다.

그림 1-22는 훈련 데이터에 강하게 부합하는 고차 다항식 수명 만족 모델의 예를 보여줍니다. 단순 선형 모델보다 훈련 데이터를 훨씬 잘 수행 할 수 있지만 실제로 예측을 신뢰합니까?

예를 들어, 국가의 이름과 같은 정보가없는 속성을 포함하여 더 많은 특성을 생활 만족 모델에 제공한다고 가정 해 보겠습니다. 이 경우 복합 모델은 이름이 aw 인 훈련 데이터의 모든 국가에서 뉴질랜드 (7.3), 노르웨이 (7.4), 스웨덴 (7.2) 및 뉴질랜드 스위스 (7.5). W 만족도 규칙이 르완다 또는 짐바브웨에 일반화되어 있다고 확신하십니까? 분명히이 패턴은 순수한 기회에 의해 훈련 데이터에서 발생했지만 모델은 패턴이 실제인지 또는 단순히 데이터의 노이즈 결과인지 여부를 알 수있는 방법이 없습니다.

경고

모델이 훈련 데이터의 양과 잡음에 비해 너무 복잡하면 초과 적합이 발생합니다. 가능한 해결 방법은 다음과 같습니다.

- 더 적은 수의 파라미터 (예 : 고차 다항식 모델이 아닌 선형 모델)를 선택하여 모델을 간소화하려면, 훈련 데이터의 속성 수를 줄이거나 모델을 구속한다.

- 더 많은 훈련 데이터를 수집한다.

- 훈련 데이터의 노이즈 감소 (예 : 데이터 오류 수정 및 이상 치 제거)

그림 1-23은 세 가지 모델을 보여줍니다. 점선은 일부 국가가없는 상태에서 훈련 된 원래 모델을 나타내고 점선은 모든 국가에서 훈련 된 두 번째 모델이며 실선은 다음과 같은 데이터로 훈련 된 선형 모델입니다. 첫 번째 모델이지만 정규화 제약 조건이 있습니다. 정규화로 인해 모델이 더 작은 기울기를 갖도록 강요 받았음을 볼 수 있습니다.이 기울기는 모델에서 훈련 된 훈련 데이터를 약간 줄이지 만 실제로는 새로운 예제로 더 잘 일반화 할 수 있습니다.
#모형을 단순하게 만들고 overfitting의 위험을 줄이는 것을 정규화라고합니다. 

학습 중에 적용 할 정규화의 양은 하이퍼 매개 변수로 제어 할 수 있습니다. 하이퍼 매개 변수는 모델이 아닌 학습 알고리즘의 매개 변수입니다. 따라서 학습 알고리즘 자체의 영향을받지 않습니다. 그것은 훈련 전에 설정되어야하며 훈련 중에 일정하게 유지되어야합니다. 정규화 하이퍼 매개 변수를 매우 큰 값으로 설정하면 거의 평평한 모델 (기울기가 0에 가까움)이 표시됩니다. 학습 알고리즘은 거의 확실하게 훈련 데이터를 과부하시키지 않지만 좋은 해결책을 찾기는 어렵습니다. 하이퍼 파라미터 튜닝은 기계 학습 시스템을 구축하는 데 중요한 부분입니다 (다음 장에서 자세한 예제를 볼 수 있습니다).

Underfitting the Training Data(교육 자료 불충분)

예상 하듯이 언더 피팅 (underfitting)은 오버 피팅의 반대입니다. 모델이 너무 단순하여 데이터의 기본 구조를 배우지 못할 때 발생합니다. 예를 들어, 삶의 만족에 대한 선형 모델은 언더 피트 (underfit) 경향이있습니다. 현실은 모형보다 훨씬 복잡하므로 훈련 예에서도 그 예측은 부정확합니다.

이 문제를 해결하기위한 주요 옵션은 다음과 같습니다.

- 더 많은 매개 변수로 더 강력한 모델 선택하기
- 학습 알고리즘에 더 나은 기능 제공 (피쳐 엔지니어링)
- 모델에 대한 제약을 감소시키는 것 (예를 들어, 정규화 하이퍼 파라미터를 감소시키는 것)

Stepping Back

이제는 머신 러닝에 대해 이미 많이 알고 있습니다. 그러나 우리는 너무 많은 개념을 통해 조금 잃어버린 느낌이 들었을 것입니다. 그래서 뒤로 물러나서 큰 그림을 봅니다.

- 기계 학습은 규칙을 명시 적으로 코딩하지 않고 데이터를 학습하여 기계가 어떤 작업을 더 잘 수행하도록하는 것입니다.

- ML 시스템에는 여러 가지 유형이 있습니다 : 감독 여부, 배치 또는 온라인, 인스턴스 기반 또는 모델 기반 등.

- ML 프로젝트에서는 교육 세트에서 데이터를 수집하고 교육 알고리즘을 학습 알고리즘에 제공합니다. 알고리즘이 모델 기반 인 경우 모델을 교육 세트에 맞추기 위해 (즉, 교육 세트 자체에 대한 좋은 예측을 위해) 일부 매개 변수를 조정 한 다음 새 사례에서도 좋은 예측을 할 수 있기를 바랍니다. 알고리즘이 인스턴스 기반 인 경우 예제를 핵심으로 학습하고 유사성 측정을 사용하여 새 인스턴스로 일반화합니다.

- 훈련 세트가 너무 작거나 데이터가 대표성을 띄지 않거나, 시끄럽다면 혹은 관련없는 기능으로 오염된 경우 시스템 성능이 좋지 않을 것 입니다. 마지막으로, 귀하의 모델은 너무 단순하지 않아야합니다 (어떤 경우에는 적합 할 것입니다). 너무 복잡하지 않아야합니다 (어떤 경우에는 적합 할 것입니다).

마지막으로 다루어야 할 중요한 주제가 하나 있습니다. 일단 모델을 교육하면 새로운 사례로 일반화되기를 "희망"만 원하지 않습니다. 당신은 그것을 평가하고, 필요하다면 그것을 미세 조정하고 싶어할 것입니다.

## Testing and Validating

모델이 새로운 케이스로 얼마나 잘 일반화되는지를 알 수있는 유일한 방법은 실제로 새로운 케이스에서 그것을 시도하는 것입니다. 이를 수행하는 한 가지 방법은 모델을 프로덕션에 배치하고 성능을 모니터링하는 것입니다. 이것은 잘 작동하지만 모델이 몹시 나쁘면 사용자가 불평 할 것입니다.

더 나은 옵션은 데이터를 교육 세트와 테스트 세트의 두 세트로 분리하는 것입니다. 이 이름이 암시 하듯이, 훈련 세트를 사용하여 모델을 훈련시키고 테스트 세트를 사용하여 모델을 테스트합니다. 새로운 경우의 오류율을 일반화 오류 (또는 표본 오류)라고하며, 테스트 세트에서 모델을 평가하면이 오류를 추정 할 수 있습니다. 이 값은 전에 보지 못한 인스턴스에서 모델이 얼마나 잘 수행되는지 알려줍니다.

교육 오류가 낮지만 (예 : 모델에서 교육 집합에 실수를 거의하지 않지만) 일반화 오류가 높으면 모델이 교육 데이터에 적합하지 않음을 의미합니다.
tip = 교육을 위해 80 %의 데이터를 사용하고 테스트를 위해 20 %를 유지하는 것이 일반적입니다.

따라서 모델을 평가하는 것만으로도 충분합니다. 테스트 세트를 사용하십시오. 이제 두 모델 (선형 모델과 다항 모델) 사이에 주저하고 있다고 가정 해보십시오. 어떻게 결정할 수 있습니까? 하나의 옵션은 두 가지를 훈련시키고 그들이 테스트 세트를 사용하여 얼마나 잘 일반화하는지 비교하는 것입니다.

검증 세트에서 너무 많은 교육 데이터를 낭비하지 않으려면 교차 검증을 사용하는 것이 일반적입니다. 교육 세트를 상보적인 하위 집합으로 분할하고 각 모델을 나머지 부분에 대해 검증하고 나머지 부분에 대해 검증합니다.모델 유형 및 하이퍼 매개 변수가 선택되면 전체 모델에서 이러한 하이퍼 매개 변수를 사용하여 최종 모델을 학습하고 일반화 된 오류를 테스트 세트에서 측정합니다.

## Exercises

이 장에서는 기계 학습에서 가장 중요한 개념을 다뤘습니다. 다음 장에서는 더 깊이 들어가서 코드를 작성하기 전에 다음 질문에 대답하는 방법을 알고 있는지 확인하십시오.

1. 머신러닝을 어떻게 정의할 것인가?

머신러닝은 데이터로부터 학습할 수 있는 시스템의 구축을 뜻합니다. 학습의 의미는 성과 평가 척도를 고려할 때 점차 나아지고 있다는 뜻입니다.


2. 네 가지 유형의 문제를 밝힐 수 있습니까?(머신러닝의 장점?)

첫째, 알고리즘 솔루션이 없는 복잡한 문제에 사용하기 좋다 
둘째, 긴 리스트를 수동으로 튜닝된 규칙으로 대체할 수 있다.
셋째, 변동하는 환경에 적응하는 시스템을 구축할 수 있다.
넷째, 인간의 학습을 돕도록 도와줍니다.

3. 라벨이 부착된 교육 세트 란 무엇입니까?

각 인스턴스마다 이상적인 솔루션이 라벨링된 교육세트라고 합니다.

4. 두 가지 가장 일반적인 감독 과제는 무엇입니까?

classification , regression 분류와 회귀 입니다.
(분류(Classification) : 입력받은 데이터가 어느 class에 속하는지 분류(classification)하는 것입니다. 
회귀(Regression) : 입력받은 데이터들로부터 수치를 예측하는 문제입니다.)

5. 네 가지 일반적인 감독되지 않은 작업은?

클러스터링/시각화/차원축소/연관 규칙학습

6. 로봇이 알려지지 않은 다양한 지형에서 걷는 것을 허용하기 위해 어떤 유형의 기계 학습 알고리즘을 사용합니까?

로봇이 알려지지 않은 지형에서 걸을 수 있도록하는 최상의 기계 학습 알고리즘은 로봇이 지형의 응답으로부터 학습하여 스스로를 최적화 할 수있는 강화된 학습입니다.
(강화학습 : 어떤 환경 안에서 정의된 에이전트가 현재의 상태를 인식하여 선택 가능한 행동들 중 보상을 최대화 하는 행동 혹은 행동순서를 선택하는 방법, 강화는 실패와 성공을 반복해가며 학습한다.)

7. 고객을 여러 그룹으로 분류하는 데 사용할 알고리즘의 유형은 무엇입니까?

고객을 여러 그룹으로 분류하는 가장 좋은 알고리즘은 감독 학습 (그룹이 레이블을 알고있는 경우) 또는 감독되지 않은 학습 (그룹 레이블이없는 경우)입니다.

8. 스팸 탐지 문제에서 감독된 학습 문제 또는 감독되지 않은 학습 문제중 어느틀을 사용할 것인가?

스팸 탐지는 레이블이 알려져 있기 때문에 감독된 학습 문제입니다 (스팸 또는 스팸 없음).

9. 온라인 학습 시스템이란 무엇입니까?

온라인 학습 시스템은 데이터가 작은 스트림으로 연속적으로 주어지면서 기계가 학습하는 학습 시스템입니다. 점진적으로 학습이 가능하여 변화하는 데이터와 자율 시스템 모두에 빠르게 적응할 수 있으며 대량의 데이터를 학습 할 수 있습니다.

10. out-of-core학습이란?

Out-of-core 학습 시스템은 컴퓨터 메모리에 맞지 않는 데이터를 처리 할 수있는 시스템입니다. 온라인 학습 시스템을 사용하여 데이터를 조금씩 공급합니다.

11. 예측을하기 위해 유사성 측정에 의존하는 학습 알고리즘 유형은 무엇입니까?

예측을 위해 유사성 측정에 의존하는 학습 알고리즘은 인스턴스 기반 알고리즘입니다.

12. 모델 매개 변수와 학습 알고리즘의 하이퍼 매개 변수의 차이점은 무엇입니까?

Model 매개 변수는 주어진 새 인스턴스에 대해 모델이 예측하는 방법을 결정합니다. 모델에는 일반적으로 둘 이상의 매개 변수 (즉, 선형 모델의 기울기)가 있습니다. 하이퍼 매개 변수는 모델이 아닌 학습 알고리즘의 매개 변수입니다. 
[모델에서는 무엇을 예측할지에 대한 결정을 하는 하나 이상의 모델 매개변수가 존재]

13. 모델 기반 학습 알고리즘은 무엇을 검색합니까? 그들이 성공하는데 사용하는 가장 일반적인 전략은 무엇입니까? 그들은 어떻게 예측을합니까?

모델 기반 학습 알고리즘은 새로운 인스턴스에 대해 최상의 결과를 제공 할 모델의 매개 변수 최적 값을 검색합니다. 함수를 최소화하기 위해 매개 변수 값이 무엇인지 결정하기 위해 종종 비용 함수 등을 사용합니다. 모델은 새 인스턴스의 값과 함수의 매개 변수를 사용하여 예측합니다.

14. Machine Learning의 주요 과제 중 네 가지를 열거 할 수 있습니까?

기계 학습의 네 가지 주요 과제는 데이터를 과도하게 적용 (너무 복잡한 모델 사용), 데이터에 부합하지 않는 모델 (간단한 모델 사용), 데이터 및 비 대표 데이터가 부족한 점입니다.

15. 모델이 교육 데이터에서 뛰어난 성능을 나타내지만 새 인스턴스가 제대로 작동하지 않는다면 어떻게됩니까? 가능한 세 가지 해결책을 말씀해 주시겠습니까?

모델이 새로운 인스턴스에 제대로 수행되지 않으면 교육 데이터에 무리가 있습니다. 이를 해결하기 위해 다음과 같은 세 가지 작업 중 하나를 수행 할 수 있습니다. 더 많은 데이터를 얻거나 간단한 모델을 구현하거나 기존 데이터 세트의 이상치나 노이즈를 제거하십시오.

16. 테스트 세트 란 무엇이며 사용하려는 이유는 무엇입니까?

테스트 세트는 훈련 데이터를 사용하여 모델을 테스트하여 성능을 확인하는 세트입니다. 모델이 얼마나 좋은지 (또는 나쁜지) 판단 할 수 있도록 테스트 세트가 필요합니다.

17. 유효성 검사 집합의 목적은 무엇입니까?

유효성 검사 집합은 서로 다른 교육 모델을 비교하는 데 사용되는 집합입니다.

18. 테스트 세트를 사용하여 하이퍼 파라미터를 조정하면 무엇이 잘못 될 수 있습니까?

테스트 세트를 사용하여 하이퍼 패러미터를 튜닝하면 모델이 특정 세트에 맞게 조정되기 때문에 샘플 밖의 데이터에서 제대로 수행되지 않을 수 있습니다.

19. 교차 유효성 검사 란 무엇이며 유효성 검사 집합을 선호하는 이유는 무엇입니까?

교차 유효성 검사는 별도의 유효성 검사 집합없이 모델을 비교하는 도구입니다. 유효성 검사 세트를 선호하는 이유는 더 많은 데이터가 필요 없으므로 유효성 검사 세트를 만들기 위해 교육 세트의 일부분을 손상시키지 않아도되기 때문입니다.
